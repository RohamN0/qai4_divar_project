```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
divar_df = pd.read_csv('../../datasets/Divar.csv')
```

```python
group1 = divar_df[divar_df['has_business_deed'] == True]['price_value'].dropna()
group2 = divar_df[divar_df['has_business_deed'] == False]['price_value'].dropna()

group1 = np.log10(group1 + 0.000001)
group2 = np.log10(group2 + 0.000001)
```

```python
group1.hist(), group2.hist()
# توزیع داده ها نرمال نیست پس از آزمون تی تست مستقل استفاده نمیکنیم
```

```python
_, p_value = stats.mannwhitneyu(group1, group2, alternative='two-sided')
if p_value < 0.05:
    print('having business deed affects the average selling price')
else:
    print('having business deed does not affects the average selling price')
```
