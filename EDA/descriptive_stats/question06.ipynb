```python
from jdatetime import datetime as shamsi
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
divar_df = pd.read_csv('../../datasets/Divar.csv')
```

```python
format = "%Y-%m-%d"
format_convert = "%Y-%m-%d %H:%M:%S"
divar_df['month_shamsi'] = divar_df['created_at_month'].apply(lambda x: shamsi.fromgregorian(datetime=datetime.strptime(x, format_convert)).strftime(format))
divar_df['month'] = divar_df['month_shamsi'].apply(lambda x: int(x.split('-')[1]))
```

```python
plt.figure(figsize=(10, 6))

df1 = divar_df.groupby('month')['rent_value'].mean()
sns.lineplot(df1, marker='o')
plt.xticks(range(1, 13))
plt.show()
```

```python
#           
plt.figure(figsize=(10, 6))

df2 = divar_df.groupby('month_shamsi')['rent_value'].mean()
sns.lineplot(df2, marker='o')
plt.xticks(rotation=90)
plt.show()
```
